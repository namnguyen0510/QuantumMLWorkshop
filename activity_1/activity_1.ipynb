The mean and variance of a random variable can be expressed as:

[**Mean**]
The mean of a random variable X is defined as:
\begin{equation} 
\mu = E[X] = \int_{-\infty}^{\infty} x f(x) dx
\end{equation}
where $\mu$ is the mean of X, $E[X]$ denotes the expected value of X, and $f(x)$ is the probability density function (PDF) of X.

[**Variance**]
The variance of a random variable X is defined as:

\begin{equation} 
\sigma^2 = Var[X] = E[(X - \mu)^2] = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx
\end{equation}

where $\sigma^2$ is the variance of X, $Var[X]$ denotes the variance of X, $\mu$ is the mean of X, and $f(x)$ is the probability density function (PDF) of X.

(*Definition*)
A vector field is a function that assigns a vector to each point in a given space, and can be expressed mathematically as:

$$\textbf{F} : \mathbb{R}^n \rightarrow \mathbb{R}^n$$

where $\textbf{F}$ is the vector field, $\mathbb{R}^n$ is the n-dimensional (**Euclidean space**), and $\rightarrow$ denotes a mapping from one space to another. The vector at each point $x \in \mathbb{R}^n$ is given by:

$$\textbf{F}(x) = \begin{bmatrix} F_1(x_1, x_2, ..., x_n) \ F_2(x_1, x_2, ..., x_n) \ \vdots \ F_n(x_1, x_2, ..., x_n) \end{bmatrix}$$
where $F_1, F_2, ..., F_n$ are the scalar functions that determine the components of the vector field.

- (**Step 2**): Define Objectives
We want to use the string code to predict one of the variable, suppose we select GC-content. Now, we have the model
$$\hat{y}_{\text{GC}} = f(\text{sgRNA})$$
We use the hat notation for the reason that $\hat{y}$ is only estimation of the actual GC-content based on analyzing the input string sgRNA.

```python
import os
from sklearn.model_selection import train_test_split
import xgboost as xgb
import optuna
from scipy import stats
import math
from optuna.samplers import TPESampler

# ENCODING METHOD
def OneHotEncoding(x):
    X = []
    for i in range(len(x)):
        if x[i] == 'A':
            X.append(np.array([1,0,0,0]))
        elif x[i] == 'T':
            X.append(np.array([0,1,0,0]))
        elif x[i] == 'G':
            X.append(np.array([0,0,1,0]))
        elif x[i] == 'C':
            X.append(np.array([0,0,0,1]))
        else:
            print('Invalid RNA sequences')
    X = np.concatenate(X, axis = 0)
    return X

def train(sgRNA, scores, dataname,
    feature_name, feature_dict, n_trials, 
    nepochs, seed, _savedir):
  savedir = 'ExpHist_{}_{}_{}_{}'.format(_savedir,dataname,feature_name, seed)
  modeldir = os.path.join(savedir,'save_models')
  try:
    os.mkdir(savedir)
    os.mkdir(modeldir)
  except:
    pass
  train_idx, test_idx = train_test_split(range(len(sgRNA)), random_state = seed)
  print(feature_dict[feature_name])
  print(train_idx)
  x_train = np.array(feature_dict[feature_name])[train_idx]
  y_train = scores[train_idx]
  x_test = np.array(feature_dict[feature_name])[test_idx]
  y_test = scores[test_idx]
  print(x_train.shape)
  print(x_test.shape)
  print(y_train.shape)
  print(y_test.shape)
  # LOAD DATA FOR xgBoost
  dtrain = xgb.DMatrix(x_train, label=y_train)
  dtest = xgb.DMatrix(x_test, label=y_test)
  evallist = [(dtrain, 'train'), (dtest, 'eval')]
  def objective(trial):
  # CREATE HYPER-PARAMETER SEARCH SPACE
    booster = trial.suggest_categorical('booster', ['gblinear', 'dart'])
    eta = trial.suggest_float('eta', 0, 1)
    max_depth = trial.suggest_int('max_depth', 1, 11)
    min_child_weight = trial.suggest_float('min_child_weight', 0, 11)
    max_delta_step = trial.suggest_float('max_delta_step', 1, 11)
    subsample = trial.suggest_float('subsample', 0, 1)
    _lambda = trial.suggest_float('lambda',0 , 5)
    alpha = trial.suggest_float('alpha',0, 5)
    grow_policy = trial.suggest_categorical('grow_policy', ['depthwise', 'lossguide'])
    param = {'booster': booster,
             'eta': eta, 
            'max_depth': max_depth,
            'min_child_weight': min_child_weight,
            'max_delta_step': max_delta_step,
            'subsample': subsample,
            'lambda': _lambda,
            'alpha': alpha,
            'grow_policy': grow_policy,
            'objective': 'reg:squarederror'}
    param['nthread'] = 12
    param['eval_metric'] = 'rmse'
    # TRAIN
    bst = xgb.train(param, dtrain, nepochs)
    y_test_pred = bst.predict(dtest)
    score = stats.spearmanr(y_test_pred, y_test, nan_policy='omit')[0]
    if math.isnan(score) == True:
      score = 0
    bst.save_model('{}/xgboost_{}__{}.json'.format(modeldir, str(trial.number+1).zfill(3), score))
    return score
  # SAVE EXP
  np.save('{}/train_idx.npy'.format(savedir), train_idx)
  np.save('{}/test_idx.npy'.format(savedir), test_idx) 
  # OPTIMIZE 
  study = optuna.create_study(sampler=TPESampler(seed = seed),study_name = savedir,
                              storage = 'sqlite:///{}/{}.db'.format(savedir,savedir),
                              load_if_exists=True,
                              direction = 'maximize')
  study.optimize(objective, n_trials=n_trials)
  return True
```

```python
# SUB-1000
data = data[:1000]
gstr = data['sgRNA']
gc_scores = data['Global_GC']
feature_dict = {}
feature_dict['OHE'] = [OneHotEncoding(g) for g in gstr]
# train argument:
# sgRNA, scores, dataname,
# feature_name, feature_dict, n_trials, 
# nepochs, seed, _savedir

train(sgRNA = gstr, scores = gc_scores, dataname = 'GFDS_26_sub1000', feature_name='OHE',feature_dict=feature_dict, 
      n_trials=1,nepochs=100, seed = 1,_savedir = 'activity_1_gfds_26_sub1000')
```
